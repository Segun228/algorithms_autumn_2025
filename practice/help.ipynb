```python
import  numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("A.csv")
```

```python
noise_cols = ['lead_time', 'avg_price_per_room']
num_cols = ["no_of_adults",	"no_of_children",	"no_of_weekend_nights",	"no_of_week_nights"]
```

```python
def add_noise(noise_cols: list, num_cols: list, df, noise_factor: float = 0.1):
    length = len(df)
    for col in noise_cols:
        if col in df.columns and pd.api.types.is_numeric_dtype(df[col]):
            ser = df[col]
            std_col = ser.std()
            if std_col > 0:
                random_vector = np.random.normal(0, std_col * noise_factor, length)
                df[col] = ser + random_vector
    for col in num_cols:
        if col in df.columns and pd.api.types.is_numeric_dtype(df[col]):
            ser = df[col]
            if pd.api.types.is_integer_dtype(ser):
                std_col = ser.std()
                if std_col > 0:
                    max_noise = max(1, int(std_col * noise_factor))
                    random_vector = np.random.randint(0, max_noise + 1, length)
                    df[col] = ser + random_vector
    return df
```

```python
df = add_noise(
    noise_cols=noise_cols,
    num_cols=num_cols,
    df = df
)
```
