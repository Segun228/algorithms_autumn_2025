```python
import random
import os
from multiprocessing import Pool, cpu_count
from tqdm import tqdm

# Настройки
ranks = list(range(2, 15))
suits = ['♠', '♥', '♦', '♣']
deck = [(rank, suit) for rank in ranks for suit in suits]

def is_pair(hand):
    return hand[0][0] == hand[1][0]

# Функция для одного процесса (не внутри if __name__ == "__main__")
def worker(args):
    n_trials, seed = args
    random.seed(seed + os.getpid())  # уникальный seed для процесса
    success = 0
    for _ in range(n_trials):
        cards = random.sample(deck, 6)
        hands = [cards[i*2:(i+1)*2] for i in range(3)]
        if all(is_pair(hand) for hand in hands):
            success += 1
    return success
```

```python
# Параметры
total_trials = 1000000000  # общее число симуляций
n_jobs = cpu_count()        # число процессов (можно уменьшить вручную)
trials_per_job = total_trials // n_jobs
remainder = total_trials % n_jobs

trial_counts = [trials_per_job] * n_jobs
trial_counts[0] += remainder  # остаток в первый процесс

# Пакуем (n_trials, seed) для каждого процесса
args = [(count, i) for i, count in enumerate(trial_counts)]

# Запуск
with Pool(processes=n_jobs) as pool:
    results = list(tqdm(pool.imap(worker, args), total=n_jobs))

# Вывод результата
total_success = sum(results)
probability = total_success / total_trials
print(f"Вероятность: {probability:.6%}")
```

```python
import random
from tqdm import tqdm

def simulate_one_run():
    total_opens = 0
    premium_count = 0

    while premium_count < 2:
        # Начинается новый лутбокс
        box = ['premium'] + ['common'] * 66
        while True:
            total_opens += 1
            reward = random.choice(box)
            if reward == 'premium':
                premium_count += 1
                break  # сбросить лутбокс
            else:
                box.remove(reward)  # удалить обычную награду

    return total_opens

# Параметры симуляции
n_runs = 10000000
results = [simulate_one_run() for _ in tqdm(range(n_runs))]

# Вывод результата
expected = sum(results) / n_runs
print(f"Математическое ожидание открытий до 2 премиумов: {expected:.4f}")
```
